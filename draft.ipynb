```python

total_distance = 500  # total travel distance in miles
average_speed = 60  # average speed in mph
discharge_rate = 0.025  # discharge rate in SOC per mile

# Helper function to calculate the charging time needed to reach a target SOC
def calculate_charging_time(current_soc, target_soc):
    # Find the current SOC and target SOC in the charging curve data
    current_soc_time = next((time for time, soc in charging_curve_data if soc >= current_soc), 60)
    target_soc_time = next((time for time, soc in charging_curve_data if soc >= target_soc), 60)

    # Return the difference in time
    return target_soc_time - current_soc_time

# Function to calculate the travel time for a charging plan
def calculate_travel_time(charging_plan):
    # Initial conditions
    current_soc = 1.0  # Start with a fully charged battery
    last_station = 0  # Start at the beginning of the route
    travel_time = 0  # Initial travel time is 0

    for station, target_soc in sorted(charging_plan.items()):
        # Calculate the energy consumption for the trip to the next charging station
        energy_consumption = (charging_station_data[station] - charging_station_data.get(last_station, 0)) * discharge_rate

        # Check if there is enough energy to reach the next charging station
        if current_soc < energy_consumption:
            return float('inf')  # The charging plan is not feasible

        # Subtract the energy consumption from the current SOC
        current_soc -= energy_consumption

        # Add the travel time to the next charging station
        travel_time += (charging_station_data[station] - charging_station_data.get(last_station, 0)) / average_speed

        # If the current SOC is less than the target SOC, charge the battery
        if current_soc < target_soc:
            charge_time = calculate_charging_time(current_soc, target_soc)
            travel_time += charge_time
            current_soc = target_soc

        last_station = station

    # Add the travel time from the last charging station to the destination
    travel_time += (total_distance - charging_station_data[last_station]) / average_speed

    return travel_time

# Genetic Algorithm parameters
population_size = 100
num_generations = 100
mutation_rate = 0.01
num_stations = len(charging_station_data)

# Create the initial population
population = [
    {station: random.random() for station in range(1, num_stations + 1)}
    for _ in range(population_size)
]

# Main loop of the Genetic Algorithm
for generation in range(num_generations):
    # Evaluate the fitness of the population
    fitness = [calculate_travel_time(plan) for plan in population]

    # Select the best individuals to reproduce
    parents = [population[i] for i in sorted(range(len(fitness)), key=lambda i: fitness[i])[:population_size // 2]]

    # Create the next generation
    population = parents + [
        {station: parents[i % len(parents)][station] for station in range(1, num_stations + 1)}
        for i in range(population_size // 2)
    ]

    # Apply mutation
    for individual in population:
        if random.random() < mutation_rate:
            individual[random.randint(1, num_stations)] = random.random()

# Print the best charging plan
best_plan = min(population, key=calculate_travel_time)
print(f"Best charging plan: {best_plan}")
print(f"Travel time: {calculate_travel_time(best_plan)}")
```

```python
import random
import numpy as np

# Given data
# distance_map: 2D array, distance between each pair of charging stations
# charging_rates: list, charging rate at each station
# travel_speed: constant, speed of the vehicle
# max_capacity: constant, maximum battery capacity of the vehicle

# Create initial population
def create_init_population(pop_size, num_stations):
    return [random.sample(range(num_stations), num_stations) for _ in range(pop_size)]

# Fitness function
def fitness(individual, distance_map, charging_rates, travel_speed, max_capacity):
    total_distance = 0
    total_charging_time = 0
    current_soc = max_capacity  # Fully charged at the start

    for i in range(len(individual) - 1):
        # Calculate distance and travel time to next station
        distance = distance_map[individual[i]][individual[i+1]]
        travel_time = distance / travel_speed
        total_distance += distance

        # Calculate energy consumption and update SOC
        energy_consumption = travel_time  # This is a simplified model
        current_soc -= energy_consumption

        if current_soc <= 0:  # Run out of battery
            return float('inf')

        # Calculate charging time at next station
        charging_time = max(10 / 60, (max_capacity - current_soc) / charging_rates[individual[i+1]])
        total_charging_time += charging_time

        # Update SOC
        current_soc = min(max_capacity, current_soc + charging_time * charging_rates[individual[i+1]])

    return total_distance / travel_speed + total_charging_time  # Total time

# Crossover function
def crossover(parent1, parent2):
    idx = random.sample(range(len(parent1)), 2)
    idx.sort()

    child1 = parent1[:idx[0]] + parent2[idx[0]:idx[1]] + parent1[idx[1]:]
    child2 = parent2[:idx[0]] + parent1[idx[0]:idx[1]] + parent2[idx[1]:]

    return child1, child2

# Mutation function
def mutate(individual):
    idx = random.sample(range(len(individual)), 2)
    individual[idx[0]], individual[idx[1]] = individual[idx[1]], individual[idx[0]]

# Genetic algorithm
def genetic_algorithm(distance_map, charging_rates, travel_speed, max_capacity, pop_size=100, num_generations=500):
    # Create initial population
    population = create_init_population(pop_size, len(distance_map))

    # Run genetic algorithm
    for _ in range(num_generations):
        # Calculate fitness
        fitness_values = [fitness(ind, distance_map, charging_rates, travel_speed, max_capacity) for ind in population]

        # Selection
        parents = [population[np.argmin(fitness_values)]] * pop_size

        # Crossover
        children = []
        for i in range(0, pop_size, 2):
            child1, child2 = crossover(parents[i], parents[i+1])
            children.extend([child1, child2])

        # Mutation
        for child in children:
            mutate(child)

        # Update population
        population = children

    # Return the best solution
    return min(population, key=lambda ind: fitness(ind, distance_map, charging_rates, travel_speed, max_capacity))

```
